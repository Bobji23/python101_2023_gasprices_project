```python
prices_master_list = []
year_prices_dictionary = {}
month_prices_dictionary = {}

#@title read_gas_prices()  (Refactor)
# Reading from the file and building the prices_master_list
def read_gas_prices(file_name = 'gasprices.txt'):
  # open the file 
  # read the line
  # split the line 
  # do any necessary data conversions 
  # build the mini list called price_mini_list
  # append this price_mini_list to the master list prices_master_list
  # update the year_prices_dictionary and month_prices_dictionary
  # return (None) out of the function
  file_contents = open(file_name, 'r')
  file_lines = file_contents.readlines()

  for s in file_lines:
    m_d_y_list = s.split('-')
    y_p_list = m_d_y_list[2].split(':')
    price = y_p_list[1].strip('\n')

    m_d_y_p_list = []
    m_d_y_p_list.append(int(m_d_y_list[0]))
    m_d_y_p_list.append(int(m_d_y_list[1]))
    m_d_y_p_list.append(int(y_p_list[0]))
    m_d_y_p_list.append(float(price))

    prices_master_list.append(m_d_y_p_list)

    # Update year_prices_dictionary
    if m_d_y_p_list[2] in year_prices_dictionary:
      year_prices_dictionary[m_d_y_p_list[2]].append(m_d_y_p_list[3])
    else:
      year_prices_dictionary[m_d_y_p_list[2]] = [m_d_y_p_list[3]]

    # Update month_prices_dictionary
    month_str = str(m_d_y_p_list[0]) + '-' + str(m_d_y_p_list[2])
    if month_str in month_prices_dictionary:
      month_prices_dictionary[month_str].append(m_d_y_p_list[3])
    else:
      month_prices_dictionary[month_str] = [m_d_y_p_list[3]]

  file_contents.close()

# Testing this function by itself
read_gas_prices()

# If the above method works correctly, then prices_master_list, year_prices_dictionary, 
# and month_prices_dictionary would have been updated with all the data from the file.
print('Number of mini lists in prices_master_list: ', len(prices_master_list))
for index in range(5):
  mini_list = prices_master_list[index]
  print(mini_list)

print('Year Prices Dictionary: ', year_prices_dictionary)
print('Month Prices Dictionary: ', month_prices_dictionary)
```
