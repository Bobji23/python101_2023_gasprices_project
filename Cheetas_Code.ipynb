```python
#Missing Dates
def get_missing_dates_report():
    import datetime

    # Read the gas prices from the file
    with open("gasprices.txt") as f:
        gas_prices = f.readlines()

    # Extract the dates from the gas prices
    dates = [datetime.datetime.strptime(gp.split(":")[0], "%m-%d-%Y").date() for gp in gas_prices]

    # Find the missing dates
    start_date = dates[0]
    end_date = dates[-1]
    all_dates = set([start_date + datetime.timedelta(days=x) for x in range((end_date - start_date).days)])
    missing_dates = sorted(list(all_dates - set(dates)))

    # Print the report
    if len(missing_dates) == 0:
        print("No missing dates found.")
    else:
        print("Missing dates:")
        for missing_date in missing_dates:
            print(missing_date.strftime("%m-%d-%Y"))
get_missing_dates_report()
```
